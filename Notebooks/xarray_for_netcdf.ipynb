```python
import xarray as xr
import os
import numpy as np
import pandas as pd
from pathlib import Path
```

```python
path = Path.cwd()/'../Data/NOAA/Atmospherical_Conditions'
air_path = path/'air.2020.nc'
uwnd_path = path/'uwnd.2020.nc'
vwnd_path = path/'vwnd.2020.nc'
rhum_path = path/'rhum.2020.nc'
pressure_values = [925,850,700,500,300,200,100]
```

```python
DS_air = xr.open_dataset(air_path)
DS_uwind = xr.open_dataset(uwnd_path)
DS_vwind = xr.open_dataset(vwnd_path)
DS_rhum = xr.open_dataset(rhum_path)
```

```python
DS_test = xr.merge(
    [DS_air.air.sel(level=pressure_values), 
     DS_uwind.uwnd.sel(level=pressure_values),
     DS_vwind.vwnd.sel(level=pressure_values),
     DS_rhum.rhum.sel(level=pressure_values)
    ])
```
